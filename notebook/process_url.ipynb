```python
from novalad import NovaladClient
```

```python
client = NovaladClient(api_key="API_KEY")
```

```python
client.run(
    url="https://abc.xyz/assets/a3/91/6d1950c148fa84c7d699abe05284/2024q4-alphabet-earnings-release.pdf"
)
```

```python
import time

while True:
    status = client.status()
    if status["status"] in ["success","failed"]:
        break
    time.sleep(30)
    print(".",end="")
print("\n\n",status)
```
