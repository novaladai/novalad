# Import NovaladClient

```python
from novalad import NovaladClient
```

to upload file from local directory using `upload` method from client and pass the file path

```python
path =r"path_to_file.pdf"

client = NovaladClient(api_key="API_KEY")

client.upload(file_path=path)
client.run()
```

you can check the status of your job using `status` method

```python
import time

while True:
    status = client.status()
    if status["status"] in ["success","failed"]:
        break
    time.sleep(30)
    print(".",end="")
print("\n\n",status)
```
